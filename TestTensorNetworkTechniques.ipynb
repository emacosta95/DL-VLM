# Results TensorNetwork time evolution

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_tdvp_200=np.load('data/itensors_calculation/data_TDVP_benchmark_200_bonddim.npz')
data_tdvp_30=np.load('data/itensors_calculation/data_TDVP_benchmark_30_bonddim.npz')
data_tdvp_50=np.load('data/itensors_calculation/data_TDVP_benchmark_50_bonddim.npz')
data_tdvp_400=np.load('data/itensors_calculation/data_TDVP_benchmark_400_bonddim.npz')
data_tedb=np.load('data/itensors_calculation/data_TEDB_benchmark.npz')
data_tdvp_more=np.load('data/itensors_calculation/data_TDVP_benchmark_400_bonddim.npz')
z_tdvp_200=data_tdvp_200['z']
z_tdvp_30=data_tdvp_30['z']
z_tdvp_50=data_tdvp_50['z']
z_tdvp_400=data_tdvp_400['z']


driving_tedb=data_tedb['driving']

z_tedb=data_tedb['z']


```

```python
fig,ax=plt.subplots(1,2,figsize=(25,10))

ax[0].plot(np.linspace(0,10,100),driving_tedb,linewidth=5,color='black')
ax[0].set_xlabel(r'tJ',fontsize=30)
ax[0].set_ylabel(r'$h(t)$',fontsize=30)
ax[0].tick_params(labelsize=30)


ax[1].plot(np.linspace(0,100,200),z_tdvp_30,label=r'bondim=30',linestyle='--',linewidth=5)
ax[1].plot(np.linspace(0,100,200),z_tdvp_50,label=r'bondim=50',linestyle='--',linewidth=5)
ax[1].plot(np.linspace(0,100,200),z_tdvp_200,label=r'bondim=200',linestyle='--',linewidth=5)
ax[1].plot(np.linspace(0,100,200),z_tdvp_400,label=r'bondim=400',linestyle='--',linewidth=5)

ax[1].plot(np.linspace(0,100,100),(z_tedb),label='TEDB',color='black',linewidth=5)
ax[1].legend(fontsize=30)
ax[1].tick_params(labelsize=30)
ax[1].set_xlabel(r'$tJ$',fontsize=30)
ax[1].set_ylabel(r'$z(t)$',fontsize=30)
plt.show()
```
