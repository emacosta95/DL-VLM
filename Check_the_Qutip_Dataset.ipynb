## Test the Quantum Annealing Simultation

#### Uniform Dataset

Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

Structure of the Driving

```python
hf=0
tf=20
rate=1
h0=5
t=np.linspace(0,tf,100)

f_t= (np.exp(1-rate*t)-1)/(np.exp(1)-1)

h=hf + (h0)* f_t
plt.plot(t,h)
plt.show()
```

Data

```python
data=np.load('data/uniform/test_for_adiabatic_uniform_size_8_tf_10.0_dt_0.001_rate_3.0_h0_5.0_hf_0.0_n_dataset_50.npz')


z=data['density']
h=data['potential']
time=data['time']
```

Visualization

```python
rate=np.linspace(0,3,50)
for i in range(0,10,1):
    plt.title(r'$\gamma=$'+f'{rate[i]}')
    plt.plot(time,z[i,:,0],label=r'$z(t)$')
    plt.plot(time,h[i,:,0],label=r'$h(t)$')
    plt.legend()
    plt.show()
```
